```python
# POS Tagging for 1-3 rows as an example
for content in data['tweet_with_stopwords'][:3]:
    text = Text(content, hint_language_code='id')  # 'id' for Indonesian
    print(f"Original Text: {content}")

    # Get POS tags for the content
    pos_tags = text.pos_tags
    print(f"POS Tags: {pos_tags}\n")

    # Count the frequency of each POS tag
    pos_count = Counter([pos for word, pos in pos_tags])
    print(f"POS Tag Count: {pos_count}\n")
```

```python
# POS Tagging for all rows in the tweet_with_stopwords column
all_pos_tags = []

for content in data['tweet_with_stopwords']:
    text = Text(content, hint_language_code='id')
    all_pos_tags.extend([pos for word, pos in text.pos_tags])

# Display the results
print(all_pos_tags)
```

```python
# Count the most common POS tags
pos_counts = Counter(all_pos_tags)
print("Most common POS tags:", pos_counts.most_common())
```

```python
import pandas as pd
from polyglot.text import Text
from collections import Counter
import re

# Menghitung semua kata dari tweet_with_stopwords
all_words = []
all_pos_tags = []

for content in data['tweet_with_stopwords']:
    # Membersihkan teks dan memisahkan kata
    words = re.findall(r'\w+', content.lower())
    all_words.extend(words)

    # Mendapatkan POS tags untuk konten
    text = Text(content, hint_language_code='id')
    pos_tags = text.pos_tags
    all_pos_tags.extend([pos for word, pos in pos_tags])

# Hitung frekuensi setiap kata
word_counts = Counter(all_words)

# Buat list untuk menyimpan informasi kata, POS, dan frekuensi
word_info = []

for word, freq in word_counts.items():
    # Mendapatkan POS tag dari kata (jika ada)
    pos_tag = None
    for original_content in data['tweet_with_stopwords']:
        text = Text(original_content, hint_language_code='id')
        for w, pos in text.pos_tags:
            if w.lower() == word:
                pos_tag = pos
                break
        if pos_tag:
            break
    word_info.append((word, pos_tag, freq))

# Membuat DataFrame dari list
word_info_df = pd.DataFrame(word_info, columns=['Kata', 'POS Tag', 'Frekuensi'])

# Mengambil 10 kata terbanyak
top_10_words = word_info_df.nlargest(10, 'Frekuensi')

# Menampilkan hasil
print(top_10_words)
```
